Como o ano de nascimento, que precisamos para calcular a idade, está em uma coluna datetime, para extraírmos o ano usaremos o comando STRFTIME. Com esse comando podemos escolher a informação a ser extraída usando os seguintes comandos:

No caso desse desafio, extraíremos somente o ano, usando o comando STRFTIME('%Y', campo_desejado). E após isso faremos uma subtração do ano atual, pegando essa informação do sistema com o comando DATE('now').

````python
%%sql
SELECT
  LastName,
  STRFTIME('%Y', BirthDate) AS ano_de_nascimento,
  STRFTIME('%Y',DATE('now')) - STRFTIME('%Y', BirthDate) AS idade
FROM
  EMPLOYEE
````
