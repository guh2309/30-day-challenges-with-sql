````python
%%sql
SELECT
  CITY,
  COUNT(DISTINCT CUSTOMERID) AS NUMERO_DE_LINHAS
FROM
  CUSTOMER
GROUP BY
  CITY
HAVING
  NUMERO_DE_LINHAS >= 2
ORDER BY
  1
````
