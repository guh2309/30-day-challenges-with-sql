Para conseguirmos encontrar os clientes com a primeira letra "A" e a terceira letra "e", usaremos o comando "SUBSTR".

Com esse comando precisamos passar três informações: o campo em análise, qual o ponto que os caracteres devem iniciar e o comprimento.

Abaixo segue uma análise do campo "FirstName", exibindo a primeira e a terceira letra.

````python
%%sql
SELECT FIRSTNAME, SUBSTR(FIRSTNAME, 1, 1) AS PRIMEIRA_LETRA, SUBSTR(FIRSTNAME, 3, 1) AS TERCEIRA_LETRA FROM CUSTOMER LIMIT 3
````

Para resolver esse desafio, a única diferença do comando anterior é que definiremos condições para essas substrings, lembrando que precisamos especificar se a letra é maiúscula ou minúscula.

````python
%%sql
SELECT FIRSTNAME FROM CUSTOMER WHERE SUBSTR(FIRSTNAME, 1, 1) = "A" AND SUBSTR(FIRSTNAME, 3, 1) = "e"
````
