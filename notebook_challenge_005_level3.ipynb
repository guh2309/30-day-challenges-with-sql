Como o SQLite não possui um comando para capitalizar o texto - deixar somente a primeira letra como maiúscula - teremos que separar a string e depois reagrupá-la.

Diferente do comando CONCAT, para concatenarmos os dados precisamos apenas usar o "||".

Para deixarmos o campo Phone no formato desejado, iremos utilizar vários REPLACE aninhados, sendo que em cada um substituiremos um caractere. Se estivessemos trabalhando no SQL Server bastaria utilizarmos o comando TRANSLATE e substituir todos os caracteres de uma só vez.

````python
%%sql
SELECT
  CustomerId AS codigo,
  (UPPER(SUBSTR(FirstName, 1, 1)) || SUBSTR(FirstName, 2) || ' ' || LOWER(LastName)) AS nome,
  LOWER(Address) AS  endereco,
  (LOWER(City) || '-' || State) AS cidade,
  UPPER(Country) AS pais,
  CAST(REPLACE(PostalCode, '-', '') AS INT) AS CEP,
  REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(Phone, '+', ''), '(', ''), ')', ''), ' ', ''), '-', '') AS telefone,
  Email
FROM
  CUSTOMER
WHERE
  Country = 'Brazil'
  AND CEP <> 0
````
